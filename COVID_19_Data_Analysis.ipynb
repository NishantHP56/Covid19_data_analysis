```python
covid = pd.read_csv('/content/COVID-19 Coronavirus.csv')
covid
```

```python
country_wise = covid.groupby('Country')['Total Cases'].sum().reset_index()
country_wise
```

```python

top_countries = country_wise.sort_values(by='Total Cases', ascending=False).head(15)

plt.figure(figsize=(10, 10))
plt.pie(top_countries['Total Cases'], labels=top_countries['Country'], startangle=140, pctdistance=0.85)
plt.title('Top 15 Countries by Total COVID-19 Cases')
plt.axis('equal')
plt.show()
```
